### Funciones

```python
def Cp(Fluid,T):
    x = T
    if Fluid=='water':
        c = [-4.84985762e-11,  8.16152621e-08, -5.48496005e-05,  1.84096211e-02, -3.08685069e+00,  2.11075200e+02]
        Cp_species = c[0]*x**5 + c[1]*x**4 + c[2]*x**3 + c[3]*T**2 + c[4]*T + c[5]   
    elif Fluid=="unused oil":
        c = [0.00428723, 0.62224249]
        Cp_species = c[0]*T + c[1]    
    return Cp_species


def Density(Fluid,T):
    x = T
    if Fluid=='water':
        c = [ 1.49227078e-08, -1.79921582e-05,  6.49530932e-03,  2.64335255e-01]
        Density_species = c[0]*x**3 + c[1]*T**2 + c[2]*T + c[3]
    elif Fluid=="unused oil":
        c = [-5.92123941e-01,  1.06126552e+03]
        Density_species = c[0]*x + c[1] 
    return Density_species


def Viscocity(Fluid,T):
    x = T
    if Fluid=='water':
        c = [ 2.86650833e-05, -3.93763073e-02,  2.03280503e+01, -4.68030316e+03, 4.06389038e+05]
        Viscocity_species = c[0]*x**4 + c[1]*x**3 + c[2]*T**2 + c[3]*T + c[4]
    elif Fluid=="unused oil":
        c = [-1.20139335e-11,  3.06384253e-08, -3.34071017e-05,  2.01894730e-02, -7.30421324e+00,  1.58203522e+03, -1.89963325e+05,  9.75619374e+06]
        Viscocity_species = c[0]*x**7 + c[1]*x**6 + c[2]*T**5 + c[3]*x**4 + c[4]*x**3 + c[5]*T**2 + c[6]*T + c[7] 
    return Viscocity_species


def Prandtl(Fluid,T):
    x = T
    if Fluid=='water':
        c = [ 2.55712532e-07, -3.50292992e-04,  1.80259651e-01, -4.13410432e+01, 3.57137220e+03]
        Pr_species = c[0]*x**4 + c[1]*x**3 + c[2]*T**2 + c[3]*T + c[4]  
    elif Fluid=="unused oil":
        c = [-1.10326004e-09,  2.82939045e-06, -3.10319059e-03,  1.88691141e+00, -6.87028683e+02,  1.49801108e+05, -1.81131299e+07,  9.37048049e+08]
        Pr_species = c[0]*x**7 + c[1]*x**6 + c[2]*T**5 + c[3]*x**4 + c[4]*x**3 + c[5]*T**2 + c[6]*T + c[7] 
    return Pr_species
```
